### QUESTION 18
&emsp;Implement a recursive function anagrams() that computes anagrams of a given word. <br>

&emsp;An anagram of a word A is a word B that can be formed by rearranging the letters of A. For example the word pot <br>
&emsp;is an anagram of the word top.<br>

&emsp;Your function will take as input the name of a file of words as well as a word. The function will print all the words<br>
&emsp;into the file that are anagrams of the input word. Name you file words.txt and use the word "spear" for your function.<br>
&emsp;(Note: include all anagrams you do not need to filter out combinations that are not actual words)<br>

&emsp;&emsp;&emsp;\>>> anagrams("word.txt", "in")

&emsp;&emsp;&emsp;ni<br>
<span style="color:blue;">Solution:</span>

```python
def anagrams(file_name, word):
    file_w = open(file_name,'w')
    if len(word) < 2:
        return word
    else:
        word_lst = []
        for i in range(0, len(word)):
            char_i = word[i]
            for j in anagrams(file_name, word[:i]+word[i+1:]):
                word_lst.append(j+char_i)
                print(j+char_i)
                file_w.writelines(j+char_i+'\n')
    return word_lst
```

```python
anagrams('words.txt','spear')
```
